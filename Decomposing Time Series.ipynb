# Differencing

```python
miles_df = pd.read_csv("us-airlines-monthly-aircraft-miles-flown.csv", header=0, parse_dates=[0])
```

```python
miles_df['lag1'] = miles_df['MilesMM'].shift(1)

# lag find krne ke liye: using .shift
```

```python
miles_df['MilesMM_diff_1'] = miles_df['MilesMM'].diff(periods=1)

# periods=1 means lag1 difference
# periods=2 means 2lag difference

# we can also directly subtract.
```

```python
miles_df.head()
```

```python
miles_df.index = miles_df['Month']
result_a = seasonal_decompose(miles_df['MilesMM'], model='additive')
result_a.plot()
```

```python
# Note:

# we are removing the trend
# here additive model is used
```

```python
# it does not give the duplicates graph
# by storing result in a variable and use plt.show()
import matplotlib.pyplot as plt

miles_df.index = miles_df['Month']
result_a = seasonal_decompose(miles_df['MilesMM'], model='additive')
fig = result_a.plot()  
plt.show()  
```

```python
miles_df.index = miles_df['Month']
result_b = seasonal_decompose(miles_df.iloc[1:,3],model = 'additive')
result_b.plot()


# this code is for removing the first nan value.
# which removes the trending.

# so we have done detrending.
```

```python
miles_df['MilesMM'].plot()
```

```python
# Above graph explanation
# -----------------------------------

# This dataset having seasonality
# becoz has every 3rd quarter the graphs is going to the peak. in years (1963 and 1964)
```

```python
# we will use periods=12 for removing seasonality
# and period=1 to removing trend
miles_df['MilesMM_diff_12'] = miles_df['MilesMM_diff_1'].diff(periods=12)
```

```python
miles_df['MilesMM_diff_12'].plot()

# Now, seasonality has been removed from this graph.
```
